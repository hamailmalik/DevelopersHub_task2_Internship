## Upload PDF

```
from google.colab import files
uploaded=files.upload()
```

## Build Vector Store

```
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
pdf=next(iter(uploaded))
docs=PyPDFLoader(pdf).load()
chunks=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50).split_documents(docs)
emb=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
vector=FAISS.from_documents(chunks,emb)
```

## Retrieval QA

```
def answer(query):
    docs=vector.similarity_search(query,k=3)
    return '\n\n'.join([d.page_content[:500] for d in docs])
print(answer('Summarize the document'))
```

## Gradio

```
import gradio as gr
iface=gr.Interface(fn=answer,inputs='text',outputs='text',title='Context-Aware Document Chatbot')
iface.launch()
```
